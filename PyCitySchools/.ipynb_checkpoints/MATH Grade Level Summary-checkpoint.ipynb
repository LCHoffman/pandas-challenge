```python
### Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
student_data = "Resources/students_complete.csv"

# Read Student Data File and store into Pandas DataFrame
student_data_df = pd.read_csv(student_data)
```

```python
#Pull out grade level, school and MATH score data
math_student_data_df = student_data_df.iloc[:, [4,3,6]]
math_student_data_df = math_student_data_df.set_index('school_name')
```

```python
#Create data frames for each grade level

#9th
math_9th_data_df = math_student_data_df.loc[(math_student_data_df["grade"] == "9th")]
math_9th_data_df = math_9th_data_df.groupby("school_name")
grade9 = round(math_9th_data_df["math_score"].mean(), 2)
grade_9_df = grade9.to_frame()

#10th
math_10th_data_df = math_student_data_df.loc[(math_student_data_df["grade"] == "10th")]
math_10th_data_df = math_10th_data_df.groupby("school_name")
grade10 = round(math_10th_data_df["math_score"].mean(), 2)
grade_10_df = grade10.to_frame()

#11th
math_11th_data_df = math_student_data_df.loc[(math_student_data_df["grade"] == "11th")]
math_11th_data_df = math_11th_data_df.groupby("school_name")
grade11 = round(math_11th_data_df["math_score"].mean(), 2)
grade_11_df = grade11.to_frame()

#12th
math_12th_data_df = math_student_data_df.loc[(math_student_data_df["grade"] == "12th")]
math_12th_data_df = math_12th_data_df.groupby("school_name")
grade12 = round(math_12th_data_df["math_score"].mean(), 2)
grade_12_df = grade12.to_frame()
```

```python
#Merge the Data Frames to form the final summary
df = pd.merge(grade_9_df, grade_10_df, on="school_name")
df =df.rename(columns={"math_score_x":"9th", "math_score_y":"10th"})
nextdf = pd.merge(df, grade_11_df, on="school_name")
nextdf = nextdf.rename(columns={"math_score":"11th"})
Math_Summary = pd.merge(nextdf, grade_12_df, on="school_name")
Math_Summary = Math_Summary.rename(columns={"math_score":"12th"})
```
