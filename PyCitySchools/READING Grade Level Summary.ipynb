```python
### Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
student_data = "Resources/students_complete.csv"

# Read Student Data File and store into Pandas DataFrame
student_data_df = pd.read_csv(student_data)
```

```python
#Pull out grade level, school and READING score data
read_student_data_df = student_data_df.iloc[:, [4,3,5]]
read_student_data_df = read_student_data_df.set_index('school_name')
```

```python
#Create data frames for each grade level

#9th
read_9th_data_df = read_student_data_df.loc[(read_student_data_df["grade"] == "9th")]
read_9th_data_df = read_9th_data_df.groupby("school_name")
grade9 = round(read_9th_data_df["reading_score"].mean(), 2)
grade_9_df = grade9.to_frame()

#10th
read_10th_data_df = read_student_data_df.loc[(read_student_data_df["grade"] == "10th")]
read_10th_data_df = read_10th_data_df.groupby("school_name")
grade10 = round(read_10th_data_df["reading_score"].mean(), 2)
grade_10_df = grade10.to_frame()

#11th
read_11th_data_df = read_student_data_df.loc[(read_student_data_df["grade"] == "11th")]
read_11th_data_df = read_11th_data_df.groupby("school_name")
grade11 = round(read_11th_data_df["reading_score"].mean(), 2)
grade_11_df = grade11.to_frame()

#12th
read_12th_data_df = read_student_data_df.loc[(read_student_data_df["grade"] == "12th")]
read_12th_data_df = read_12th_data_df.groupby("school_name")
grade12 = round(read_12th_data_df["reading_score"].mean(), 2)
grade_12_df = grade12.to_frame()
```

```python
#Merge the Data Frames to form the final summary
df = pd.merge(grade_9_df, grade_10_df, on="school_name")
df =df.rename(columns={"reading_score_x":"9th", "reading_score_y":"10th"})
nextdf = pd.merge(df, grade_11_df, on="school_name")
nextdf = nextdf.rename(columns={"reading_score":"11th"})
Reading_Summary = pd.merge(nextdf, grade_12_df, on="school_name")
Reading_Summary = Reading_Summary.rename(columns={"reading_score":"12th"})
```
